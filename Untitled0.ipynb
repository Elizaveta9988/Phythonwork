```python
from sympy import*
x= Symbol("x")
fx=x**3-50*x
gx=-1*x**4+88*x**2-241
fgx=fx-gx
```

```python
roots=solve(Eq(fx,gx))
roots
```

```python
cross_dict={}
c=1
for z in roots:
    cross_dict[c]={"x": round(N(z),5),"y": fx.evalf(5,subs={x:z}, chop=True)}
    c+=1
cross_dict
```

```python
cross_dict
```

```python
print(*[(cross_dict[i]["x"], cross_dict[i]["y"]) for i in cross_dict.keys()], sep=",")
```
